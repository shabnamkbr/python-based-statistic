```python
import pandas as pd                          #to read exel
from scipy.stats import shapiro              #Shapiro-Wilk for Normal Distribution

    

!pip install openpyxl


df = pd.read_excel(r"C:\kar\matlab\subsub.xlsx")              #read exel and save in df variable

stat, p = shapiro(df["FSIQ"].dropna())                       #check the FSIQ 

print("Statistic =", stat)
print("p-value =", p)

if p > 0.05:
    print("data in normal")
else:
    print("not normal")
    
```

```python
from scipy.stats import ttest_ind                                  #indipendent t test

male = df[df["Gender"] == "M"]["FSIQ"].dropna()                     #pick males
female = df[df["Gender"] == "F"]["FSIQ"].dropna()                   #females

print("Number of men:", len(male))
print("Number of women:", len(female))

t, p = ttest_ind(male, female)

print("t =", t)
print("p =", p)

print("\n this shows:")
if p > 0.05:
    print("No meaningful dif between men and women about IQ")
else:
    print("meaningful dif between men and women about IQ")
```

```python
from scipy.stats import pearsonr                       #Pearson for corolaton between age and IQ

r, p = pearsonr(df["Age"], df["FSIQ"])

print("Correlation =", r)                           # r is between -1 and  1
print("p =", p)

print("\n shows:")

if p > 0.05:
    print("no meaningful connection between age and iq.")
else:
    if r > 0:
        print("posetive meaningful connection between age and iq.")
    else:
        print("negative meaningful connection between age and iq.")
```

```python
import pandas as pd                                          #Chi-square for IQ and fe/m
from scipy.stats import chi2_contingency

median = df["FSIQ"].median()                                 # IQ's median

df["IQ_Group"] = df["FSIQ"].apply(
    lambda x: "Low" if x < median else "High"                  # lower than median vs higher than that
)

table = pd.crosstab(df["IQ_Group"], df["Gender"])            #Frequency table (male and high iq - female and high iq - male and low iq....)

chi2, p, dof, expected = chi2_contingency(table)

print(table)
print("Chi2 =", chi2)
print("p =", p)

print("\n shows")

if p > 0.05:
    print("No meaningful connection between fe/m and IQ.")
else:
    print("meaningful connection between fe/m and IQ.")
```

```python
low = df[df["IQ_Group"] == "Low"]["Age"]              #low IQs chosen to show age
high = df[df["IQ_Group"] == "High"]["Age"]            #high IQ

from scipy.stats import ttest_ind

t, p = ttest_ind(low, high)                          # is there a difference between the ages of high iq and low iq? --> t test

print(t, p)

print("\n shows")

if p > 0.05:
    print("No meaningful connection between age and IQ.")
else:
    print("meaningful connection between age and IQ.")

```
